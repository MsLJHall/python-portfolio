```python
# f-strings - combining variables into formatted text
tool_name = "AI Readiness Assessment"
version = "1.0"
author = "Ms. LJ Hall, Ph.D., MAOL"
purpose = "Evaluate organizational readiness for AI adoption"

header = f"Tool: {tool_name} v{version} | Created by: {author}"
print(header)

description = f"Purpose: {purpose}"
print(description)

print(f"{'=' * 50         }")
print(f"  {tool_name}")
print(f"  Version {version}")
print(f"  Developed by {author}")
print(f"{'=' * 50}")
```

This tool collects information about an organization to assess their AI readiness.

```python
# Collect organizational information
print("=" * 50)
print("  ORGANIZATIONAL AI READINESS ASSESSMENT")
print("=" * 50)
print()
print("Please answer the following questions about your organization.")
print()

org_name = input("Organization name: ")
industry = input("Industry (e.g., Healthcare, Finance, Education): ")
employees = input("Number of employees: ")
challenge = input("Biggest operational challenge right now: ")
uses_ai = input("Does your organization currently use any AI tools? (yes/no): ")
```

```python
# Display collected organizational information
print("\n--- Organizational Information Summary ---")
print(f"Organization Name: {org_name}")
print(f"Industry: {industry}")
print(f"Number of Employees: {employees}")
print(f"Biggest Operational Challenge: {challenge}")
print(f"Currently Uses AI Tools: {uses_ai}")
print("--------------------------------------")
```

```python
# Build the executive summary report
report = f"""
{'=' * 50}
  AI READINESS ASSESSMENT - EXECUTIVE SUMMARY
{'=' * 50}

ORGANIZATION PROFILE
--------------------
Organization: {org_name}
Industry:     {industry}
Size:         {employees} employees
AI Status:    {"Currently using AI tools" if uses_ai == "yes" else "Not yet using AI tools"}

INITIAL ASSESSMENT NOTES
------------------------
Primary Challenge: {challenge}

This organization operates in the {industry} sector
with {employees} team members. Their primary operational
challenge is: {challenge}.

{"They have begun their AI journey and are already leveraging AI tools." if uses_ai == "yes" else "They have not yet adopted AI tools, presenting a greenfield opportunity for strategic AI implementation."}

NEXT STEPS
----------
- Schedule a detailed AI capability assessment
- Identify quick-win AI use cases for {industry}
- Develop an AI roadmap aligned with solving: {challenge}

{'=' * 50}
Report generated by: {tool_name} v{version}
Strategist: {author}
{'=' * 50}
"""

print(report)
```

## How This Code Works

- **Variables** (`org_name`, `industry`, etc.) store the data collected from the user.
- **`input()`** displays a prompt and waits for the user to type a response. Everything entered is stored as text (a string).
- **f-strings** (`f"..."`) let us insert variable values directly into formatted text using curly braces `{}`.

```python
# AI Maturity Stage Assessment - Deeper Questions
print("=" * 50)
print("  AI MATURITY STAGE ASSESSMENT")
print("=" * 50)
print()
print("Let's dig deeper to determine your AI maturity stage.")
print()

data_rating = input("How would you rate your data infrastructure? (1-5): ")
has_skills = input("Does your team have AI/ML skills? (yes/no): ")
has_governance = input("Do you have an AI governance policy? (yes/no): ")
```

```python
# Calculate maturity score
score = 0

if uses_ai == "yes":
    score = score + 1
if has_skills == "yes":
    score = score + 1
if has_governance == "yes":
    score = score + 1
if int(data_rating) >= 4:
    score = score + 1

# Determine maturity stage
if score >= 3:
    stage = "Stage 3: AI Ready"
    recommendation = "Your organization is positioned to scale AI initiatives. Focus on governance and measurable outcomes."
elif score >= 1:
    stage = "Stage 2: AI Curious"
    recommendation = "Your organization has foundational elements in place. Prioritize data infrastructure and team upskilling."
else:
    stage = "Stage 1: AI Aware"
    recommendation = "Your organization is at the beginning of the AI journey. Start with education and identifying low-risk pilot projects."

# Print the maturity report
print()
print("=" * 50)
print(f"  {stage.upper()}")
print("=" * 50)
print()
print(f"Organization: {org_name}")
print(f"Maturity Score: {score}/4")
print(f"Stage: {stage}")
print()
print("RECOMMENDATION")
print("-" * 30)
print(recommendation)
print()
print("SCORING BREAKDOWN")
print("-" * 30)
print(f"  AI Tools in Use:      {'Yes' if uses_ai == 'yes' else 'No'}")
print(f"  AI/ML Team Skills:    {'Yes' if has_skills == 'yes' else 'No'}")
print(f"  Governance Policy:    {'Yes' if has_governance == 'yes' else 'No'}")
print(f"  Data Infrastructure:  {data_rating}/5")
print()
print("=" * 50)
print(f"Assessment by: {author}")
print("=" * 50)
```
