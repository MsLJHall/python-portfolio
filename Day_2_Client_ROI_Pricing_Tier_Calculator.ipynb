This tool automates client pricing decisions. It collects project details, calculates ROI from operational efficiencies, and assigns clients to volume discount tiers automatically.

## Client Input Dashboard

Enter your client's project details in the form fields on the right. These values will be used to calculate ROI and determine their pricing tier.

```python
#@title Client ROI & Pricing Tier Calculator
#@markdown Enter your client's details below.

client_name = 'Acme Corp' #@param {type:"string"}
project_value_text = '5000' #@param {type:"string"}
monthly_ops_spend_text = '8000' #@param {type:"string"}

print(f"Client data received for: {client_name}")
print(f"Project value (text): {project_value_text}")
print(f"Monthly spend (text): {monthly_ops_spend_text}")
print()
print(f"Type of project_value_text: {type(project_value_text)}")
print(f"Type of monthly_ops_spend_text: {type(monthly_ops_spend_text)}")
```

```python
# Type casting - converting text inputs to math-ready numbers
project_value = float(project_value_text)
monthly_ops_spend = float(monthly_ops_spend_text)

print(f"project_value: {project_value}")
print(f"Type after casting: {type(project_value)}")
print()
print(f"monthly_ops_spend: {monthly_ops_spend}")
print(f"Type after casting: {type(monthly_ops_spend)}")
print()
```

```python
# Why casting matters - string vs number multiplication
print("WITHOUT casting (string * 2):")
print(f"  '{project_value_text}' * 2 = '{project_value_text * 2}'")
print()
print("WITH casting (float * 2):")
print(f"  {project_value} * 2 = {project_value * 2}")
```

```python
# ROI Calculations
annual_ops_spend = monthly_ops_spend * 12
efficiency_rate = 0.30  # 30% operational savings from process improvements
annual_savings = annual_ops_spend * efficiency_rate
roi_percentage = (annual_savings / project_value) * 100
payback_months = project_value / (annual_savings / 12)
```

```python
def assign_pricing_tier(project_value, monthly_ops_spend):
    """
    Assigns a pricing tier and associated benefits based on project value and monthly operational spend.

    Args:
        project_value (float): The total value of the client's project.
        monthly_ops_spend (float): The client's monthly operational spend.

    Returns:
        tuple: A tuple containing (tier, discount, tier_benefit).
    """
    if project_value >= 50000 and monthly_ops_spend >= 10000:
        tier = "Platinum"
        discount = 0.20
        tier_benefit = "Priority support + 20% volume discount"
    elif project_value >= 20000 and monthly_ops_spend >= 5000:
        tier = "Gold"
        discount = 0.10
        tier_benefit = "Dedicated account manager + 10% volume discount"
    else:
        tier = "Silver"
        discount = 0.05
        tier_benefit = "Standard support + 5% volume discount"
    return tier, discount, tier_benefit
```

### Calling the Pricing Tier Function

```python
# Assign pricing tier using the new function
tier, discount, tier_benefit = assign_pricing_tier(project_value, monthly_ops_spend)
```

```python
# Financial Summary Report
print(f"{'=' * 50}")
print(f"  CLIENT ROI ANALYSIS")
print(f"{'=' * 50}")
print()
print(f"Client: {client_name}")
print(f"Project Investment: ${project_value:,.2f}")
print(f"Monthly Operational Spend: ${monthly_ops_spend:,.2f}")
print()
print(f"PROJECTED RETURNS")
print(f"{'-' * 30}")
print(f"Annual Operational Spend: ${annual_ops_spend:,.2f}")
print(f"Estimated Annual Savings (30%): ${annual_savings:,.2f}")
print(f"ROI: {roi_percentage:.1f}%")
print(f"Payback Period: {int(payback_months)} months")
print(f"{'=' * 50}")
```

```python
# This cell previously contained the pricing tier assignment logic. It has been refactored into the assign_pricing_tier function.
```

```python
discounted_price = project_value * (1 - discount)
savings_from_discount = project_value - discounted_price

# Pricing Tier Report
print()
print(f"{'=' * 50}")
print(f"  PRICING TIER RECOMMENDATION")
print(f"{'=' * 50}")
print()
print(f"Client: {client_name}")
print(f"Assigned Tier: {tier}")
print(f"Benefits: {tier_benefit}")
print()
print(f"PRICING BREAKDOWN")
print(f"{'-' * 30}")
print(f"Standard Project Price: ${project_value:,.2f}")
print(f"Volume Discount ({int(discount * 100)}%): -${savings_from_discount:,.2f}")
print(f"Final Client Price: ${discounted_price:,.2f}")
print()
print(f"{'=' * 50}")
print(f"Generated by: Client ROI & Pricing Tier Calculator v1.0")
print(f"{'=' * 50}")
```

## How This Code Works

- **Type Casting** (`float()`) converts text form inputs into decimal numbers that Python can do math with.
- **Math Operators** (`*`, `/`, `-`) calculate annual spend, savings, ROI percentages, and discount amounts.
- **Conditional Logic** (`if/elif/else`) routes clients to different pricing tiers based on their project value **and** monthly operational spend.
- **f-strings** with format specifiers (`:,.2f`) add dollar signs, commas, and decimal places for professional output.

To run pytest, we'll first save the `assign_pricing_tier` function and its tests into a temporary Python file. This allows `pytest` to discover and execute the tests properly.

```python
%%writefile test_pricing_tier.py

def assign_pricing_tier(project_value, monthly_ops_spend):
    """
    Assigns a pricing tier and associated benefits based on project value and monthly operational spend.

    Args:
        project_value (float): The total value of the client's project.
        monthly_ops_spend (float): The client's monthly operational spend.

    Returns:
        tuple: A tuple containing (tier, discount, tier_benefit).
    """
    if project_value >= 50000 and monthly_ops_spend >= 10000:
        tier = "Platinum"
        discount = 0.20
        tier_benefit = "Priority support + 20% volume discount"
    elif project_value >= 20000 and monthly_ops_spend >= 5000:
        tier = "Gold"
        discount = 0.10
        tier_benefit = "Dedicated account manager + 10% volume discount"
    else:
        tier = "Silver"
        discount = 0.05
        tier_benefit = "Standard support + 5% volume discount"
    return tier, discount, tier_benefit


import pytest

@pytest.mark.parametrize(
    "project_value, monthly_ops_spend, expected_tier, expected_discount, expected_benefit",
    [
        (50000.0, 10000.0, "Platinum", 0.20, "Priority support + 20% volume discount"), # Boundary Test 1: Platinum Threshold (Exact)
        (49999.0, 10000.0, "Gold", 0.10, "Dedicated account manager + 10% volume discount"), # Boundary Test 2: Just Below Platinum PV
        (50000.0, 9999.0, "Gold", 0.10, "Dedicated account manager + 10% volume discount"),  # Boundary Test 3: Just Below Platinum MS
        (20000.0, 5000.0, "Gold", 0.10, "Dedicated account manager + 10% volume discount"),  # Boundary Test 4: Gold Threshold (Exact)
        (19999.0, 5000.0, "Silver", 0.05, "Standard support + 5% volume discount"),  # Boundary Test 5: Just Below Gold PV
        (20000.0, 4999.0, "Silver", 0.05, "Standard support + 5% volume discount"),  # Boundary Test 6: Just Below Gold MS
        (1.0, 1.0, "Silver", 0.05, "Standard support + 5% volume discount") # Boundary Test 7: Minimum Values (Silver)
    ]
)
def test_assign_pricing_tier_boundary_conditions(
    project_value, monthly_ops_spend, expected_tier, expected_discount, expected_benefit
):
    tier, discount, benefit = assign_pricing_tier(project_value, monthly_ops_spend)
    assert tier == expected_tier
    assert discount == expected_discount
    assert benefit == expected_benefit
```
