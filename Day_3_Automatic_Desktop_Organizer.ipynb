## Messy Desktop Simulation

The next code cell generates 25 realistic business files in a single folder to simulate a cluttered desktop. Click the play button to create the mess, then watch the organizer sort it.

```python
#@title Simulate a Messy Desktop
#@markdown Click the play button to create 25 business files.

import os

# Create the messy desktop folder
desktop_path = "/content/messy_desktop"
os.makedirs(desktop_path, exist_ok=True)

# 25 realistic business files
business_files = [
    "Q1_Revenue_Report.pdf",
    "Client_Invoice_042.pdf",
    "Annual_Budget_Summary.pdf",
    "Employee_Handbook.pdf",
    "Meeting_Minutes_June.pdf",
    "Marketing_Budget_2025.xlsx",
    "Sales_Forecast_Q2.xlsx",
    "Inventory_Tracker.xlsx",
    "Client_Contact_List.csv",
    "Expense_Report_May.csv",
    "Team_Photo_Retreat.png",
    "Product_Screenshot.png",
    "Office_Floor_Plan.png",
    "Company_Logo_Final.jpg",
    "Event_Banner_Draft.jpg",
    "Board_Meeting_Deck.pptx",
    "Sales_Pitch_ClientX.pptx",
    "Onboarding_Training.pptx",
    "Project_Proposal_Draft.docx",
    "Contract_Template.docx",
    "Weekly_Status_Update.docx",
    "Vendor_Agreement.docx",
    "Quick_Notes.txt",
    "Spin_That.PDF"
    "YouTube_Friends.docx"
    "Password_Reset_Instructions.txt",
    "Server_Maintenance_Log.txt",
]

# Create all 27 files
for filename in business_files:
    filepath = os.path.join(desktop_path, filename)
    open(filepath, "w").close()
    print(f"Created {len(business_files)} files in {desktop_path}")
print()

# Display the "before" snapshot
print(f"{'=' * 50}")
print(f"  MESSY DESKTOP - BEFORE ORGANIZING")
print(f"{'=' * 50}")
print()

files = sorted(os.listdir(desktop_path))
for f in files:
    print(f"  {f}")

print()
print(f"Total files: {len(files)}")
print(f"{'=' * 50}")
```

```python
#@title Run the Automatic Desktop Organizer
#@markdown Click play to sort all 25 files into categorized folders.

import shutil

# Dictionary mapping file extensions to folder categories
file_categories = {
    ".pdf": "Reports",
    ".xlsx": "Spreadsheets",
    ".csv": "Spreadsheets",
    ".png": "Images",
    ".jpg": "Images",
    ".pptx": "Presentations",
    ".docx": "Documents",
    ".txt": "Documents",
}
```

```python
# Organize files into category folders
files_moved = 0

for filename in os.listdir(desktop_path):
    filepath = os.path.join(desktop_path, filename)

    # Skip directories - only process files
    if not os.path.isfile(filepath):
        continue

    # Get the file extension (lowercase for consistent matching)
    name, extension = os.path.splitext(filename)
    extension = extension.lower()

    # Look up the category in the dictionary
    if extension in file_categories:
        category = file_categories[extension]
    else:
        category = "Other"

    # Create the category folder if it doesn't exist
    category_path = os.path.join(desktop_path, category)
    os.makedirs(category_path, exist_ok=True)

    # Move the file into the category folder
    destination = os.path.join(category_path, filename)
    shutil.move(filepath, destination)
    files_moved += 1

print(f"Organizing complete! Moved {files_moved} files.")
```

```python
#@title View Your Organized Desktop
#@markdown Click play to see the clean folder structure.

print(f"{'=' * 50}")
print(f"  ORGANIZED DESKTOP - AFTER")
print(f"{'=' * 50}")
print()

for folder_name in sorted(os.listdir(desktop_path)):
    folder_path = os.path.join(desktop_path, folder_name)
    if os.path.isdir(folder_path):
        folder_files = sorted(os.listdir(folder_path))
        print(f"  {folder_name}/ ({len(folder_files)} files)")
        for f in folder_files:
            print(f"    {f}")
        print()

print(f"{'=' * 50}")
print(f"Desktop organized successfully!")
print(f"{'=' * 50}")
```

## How This Code Works

- **Dictionaries** map file extensions to folder category names, providing instant key-value lookups without chains of if/elif statements.
- **For Loops** iterate through every file in the messy desktop folder, processing each one automatically.
- **`os.path.splitext()`** splits a filename into its name and extension (e.g., `"report.pdf"` becomes `("report", ".pdf")`).
- **`os.makedirs(exist_ok=True)`** creates destination folders only when they don't already exist, preventing errors on repeat runs.
- **`shutil.move()`** physically relocates files from the flat folder into their categorized subfolders.
- **`os.path.isfile()`** checks whether a path points to a file (not a folder), so the loop skips directories.
- **`.lower()`** converts extensions to lowercase for case-insensitive matching (handles .PDF, .Xlsx, etc.).
